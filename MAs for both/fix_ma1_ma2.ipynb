```python
import pandas as pd
```

```python
cit = pd.read_csv('AD_citationsConcepts.csv', sep=';')
ref = pd.read_csv('AD_refConcepts.csv', sep=';')
ma1 = pd.read_csv('MA1_Data.csv', sep=';')
ma2 = pd.read_csv('MA2_Data.csv', sep=';')
```

```python
df_inner_ma1 = pd.merge(cit, ma1, on='pmid', how='inner')
df_inner_ma11 = pd.merge(ref, df_inner_ma1, on='pmid', how='inner')
df_inner_ma11 = df_inner_ma11[~df_inner_ma11.cuis_x.str.startswith(('[]'))]
df_inner_ma11 = df_inner_ma11[~df_inner_ma11.cuis_y.str.startswith(('[]'))]
df_inner_ma11
```

```python
df_inner_ma2 = pd.merge(cit, ma2, on='pmid', how='inner')
df_inner_ma22 = pd.merge(ref, df_inner_ma2, on='pmid', how='inner')
df_inner_ma22 = df_inner_ma22[~df_inner_ma22.cuis_x.str.startswith(('[]'))]
df_inner_ma22 = df_inner_ma22[~df_inner_ma22.cuis_y.str.startswith(('[]'))]
df_inner_ma22
```

```python
df_inner_ma11 = df_inner_ma11[['pmid', 'Manual Class']]
```

```python
df_inner_ma22 = df_inner_ma22[['pmid', 'Manual Class']]
```

```python
df_inner_ma11.to_csv('MA1_Data1.csv', sep=';', index=False)
```

```python
df_inner_ma22.to_csv('MA2_Data2.csv', sep=';', index=False)
```

```python
ma1txt = pd.read_csv('MA1_pmids.txt', header=None)
ma1txt.columns = ["pmid"]
df = pd.merge(df_inner_ma11, ma1txt, on='pmid', how='inner')
df = df[['pmid']]
df.to_csv('MA1_pmids1.txt', index=False, header=None)
```

```python
ma2txt = pd.read_csv('MA2_pmids.txt', header=None)
ma2txt.columns = ["pmid"]
df2 = pd.merge(df_inner_ma22, ma2txt, on='pmid', how='inner')
df2 = df2[['pmid']]
df2.to_csv('MA2_pmids2.txt', index=False, header=None)
```
